```python
import pandas as pd
data = {
    "Student": ["Rahul","Priya","Akash","Sneha","Rohan","Anika","Sameer","Meera","Karan","Tanya"],
    "Math": [85,90,70,95,60,88,72,82,78,91],
    "Physics": [78,88,65,92,75,80,68,90,85,95],
    "Chemistry": [92,79,80,89,68,85,74,88,80,94]
}
```

```python
df = pd.DataFrame(data)
print(df)
```
