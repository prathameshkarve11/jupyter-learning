```python
import pandas as pd 
import numpy as np

data = {
    "Movie": ["Inception", "Interstellar", "The Matrix", "Parasite", "Avengers: Endgame"],
    "Rating": [8.8, 8.6, 8.7, 8.6, 8.4],
    "Year": [2010, 2014, 1999, 2019, 2019]
}

df = pd.DataFrame(data)
print(df)
```

```python
sorted_df = df.sort_values(by = "Rating" , ascending = False)
print(sorted_df)
```

```python
yearsorted_df = df.sort_values(by="Year")
print(yearsorted_df)
```

```python
moviesnamesorted_df = df.sort_values(by="Movie")
print(moviesnamesorted_df)
```
